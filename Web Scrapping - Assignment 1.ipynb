# 2) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
import pandas as pd
```

```python
page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
```

```python
page
```

```python
Soup = BeautifulSoup(page.content)
```

```python
Soup
```

```python
url = "https://presidentofindia.nic.in/former-presidents.htm"
```

```python
name = []
for i in Soup.find_all('h3'):
    name.append(i.text)
    
```

```python
import re
of = []
for i in Soup.find_all('div',class_="presidentListing"):
    x = i.text.replace('\n','').split(')')[1].replace('Term of Office:','')
    y = re.sub(r"http\S+", "", x)
    of.append(y)
```

```python
president = pd.DataFrame({'Former Presidents':name, 'Term of Office':of})
print('Former Presidents List')
president
```

# 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame
i) Headline
ii) Time
iii) News Link

```python
News = requests.get('https://www.cnbc.com/world/?region=world')
News
```

```python
soup = BeautifulSoup(News.content, 'html.parser')
```

```python
headline = []
for i in soup.find_all('div',class_="RiverPlusCard-container"):
    headline.append(i.text.split('•')[0])
```

```python
Newslink = []
for i in soup.find_all('div',class_="RiverPlusCard-container"):
    sub = i.find('a')
    Newslink.append(sub.get('href'))
```

```python
CNB = pd.DataFrame({'Headlines,Time':headline,'Links':Newslink})
CNB
```

# 7) Write a python program to scrape mentioned details from dineout.co.inand make data frame
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
dine = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
dine
```

```python
soup = BeautifulSoup(dine.content)
```

```python
name = []
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    name.append(i.text)
```

```python
cuisine = []
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
```

```python
location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
```

```python
rating = []
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
image = []
```

```python
image = []
for i in soup.find_all('img',class_="no-img"):
    image.append(i['data-src'])
```

```python
dineout = pd.DataFrame({'Restaurant Name':name, 'Cuisine':cuisine, 'Location':location, 'Rating':rating, 'Image URL':image})
```
