4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand
Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the
details is missing then replace it by “- “. Save your results in a dataframe and CSV. 

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.flipkart.com/')
```

```python
close = driver.find_element(By.XPATH,'/html/body/div[2]/div/div/button')
close.click()
```

```python
search = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search.send_keys('oneplus nord')
search.submit()
```

```python
Brand = []
Name =[]
Color=[]
for i in driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]'):
    Brand.append(i.text.split(' ')[0])
    Name.append(i.text)
    co = i.text.split('(')[-1]
    Color.append(co.split(',')[0])
```

```python
RAM =[]
ROM =[]
Primary_Camera =[]
Display=[]
Battery =[]

for i in driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]'):
    RA = i.text.split('\n')[0]
    RAM.append(RA.split('|')[0])
    ROM.append(RA.split('|')[1])
    
    Display.append(i.text.split('\n')[1])
    Battery.append(i.text.split('\n')[-2])
    
    Camera = i.text.split('\n')[2]
    Primary_Camera.append(Camera.split('|')[0])
```

```python
Price = []
Url=[]
for i in driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]'):
    Price.append(i.text)
for i in driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]'):
    Url.append(i.get_attribute('href'))
```

```python
print(len(Brand),len(Name),len(Color),len(RAM),len(ROM),len(Primary_Camera),len(Display),len(Battery),len(Price),len(Url))
```

```python
Phone = pd.DataFrame({'Brand Name':Brand,'Phone Name':Name,'Colour':Color,'Phone RAM':RAM,'Phone ROM':ROM,'Primary Camera':Primary_Camera,'Phone Display':Display,'Phone Battery':Battery,'Price':Price,'Url':Url})
```

```python
Phone.to_csv('Phone.csv')
```

```python
pd.read_csv('Phone.csv')
```

```python
driver.close()
```
